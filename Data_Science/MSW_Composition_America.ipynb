```python
#Waste_Generated = pd.DataFrame(country_data_df, columns =[
    
waste = pd.DataFrame(country_data_df, columns=['country_name', 
                                'income_id',
                                'gdp', 
                                'population_population_number_of_people',
                                'composition_food_organic_waste_percent', 
                                'composition_glass_percent',
                                'composition_metal_percent', 
                                'composition_other_percent',
                                'composition_paper_cardboard_percent', 
                                'composition_plastic_percent',
                                'composition_rubber_leather_percent', 
                                'composition_wood_percent',
                                'composition_yard_garden_green_waste_percent',
                                'special_waste_agricultural_waste_tons_year',
                                'special_waste_construction_and_demolition_waste_tons_year',
                                'special_waste_e_waste_tons_year',
                                'special_waste_hazardous_waste_tons_year',
                                'special_waste_industrial_waste_tons_year',
                                'special_waste_medical_waste_tons_year',
                                'total_msw_total_msw_generated_tons_year'])

waste.head()
```

```python
All_waste = waste.rename(columns={'country_name': 'Country', 
                                'income_id': "Income",
                                'gdp': 'GDP', 
                                'population_population_number_of_people': 'Population',
                                'composition_food_organic_waste_percent': 'Organic', 
                                'composition_glass_percent': 'Glass',
                                'composition_metal_percent': 'Metal', 
                                'composition_other_percent': 'Other',
                                'composition_paper_cardboard_percent': 'Cardboard', 
                                'composition_plastic_percent': 'Plastic',
                                'composition_rubber_leather_percent': 'Rubber_Leather', 
                                'composition_wood_percent': 'Wood',
                                'composition_yard_garden_green_waste_percent': 'Yard_Garden',
                                'special_waste_agricultural_waste_tons_year': 'Agricultural',
                                'special_waste_construction_and_demolition_waste_tons_year': 'Building',
                                'special_waste_e_waste_tons_year': 'Electronic',
                                'special_waste_hazardous_waste_tons_year': 'Hazardous',
                                'special_waste_industrial_waste_tons_year': 'Industrial',
                                'special_waste_medical_waste_tons_year': 'Medical',
                                'total_msw_total_msw_generated_tons_year': 'Total_MSW'})

All_waste.fillna(0).head(3)
```

```python
General_waste = All_waste[['Country', 
                       'Income', 
                       'Organic', 
                        'Glass', 
                        'Metal', 
                        'Other', 
                        'Cardboard', 
                        'Plastic', 
                        'Rubber_Leather', 
                        'Wood', 
                        'Yard_Garden'
                        ]]
General_waste.fillna(0).set_index('Country').head(3)
```

```python
# waste_generated_LIC = df_Africa.groupby('Income').get_group('UMC').fillna(0).drop_duplicates().reset_index('Country')
America_summary = df_Americas.sum(axis=0).fillna(0)
America_summary
```

```python
a =  pd.DataFrame(America_summary)
a.reset_index()
a.rename(columns={0:'Percent'}, inplace=True)
pd.options.display.float_format = '{:,.0f}'.format
a.index.names = ['Waste']
a.reset_index()
```
