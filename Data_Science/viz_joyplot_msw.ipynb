```python
%matplotlib notebook
import pandas as pd
#from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
country_data = "/Users/shilpa/Desktop/Boringppl_project/WasteManagement/Data/country_level_data_0.csv"
country_data_df = pd.read_csv(country_data, encoding = 'utf-8')
#country_data_df.head(1)
```

```python
Total_MSW = country_data_df[['country_name', 
                       'income_id', 
                       'gdp', 
                       'population_population_number_of_people', 
                       'total_msw_total_msw_generated_tons_year']]
Total_MSW.fillna(0).set_index('country_name').head(3)
```

```python
# rate of MSW generation per day = total_msw_total_msw_generated_tons_year / 365

Total_MSW['msw_per_day'] =  Total_MSW['total_msw_total_msw_generated_tons_year'] / 365
Total_MSW['msw_per_day_per_person'] = Total_MSW['total_msw_total_msw_generated_tons_year']\
                                        /Total_MSW["population_population_number_of_people"]
```

```python
Total_MSW['default_rank'] = Total_MSW['msw_per_day_per_person']
Total_MSW.set_index('country_name')
```

```python
import joypy
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
%matplotlib inline

# Draw Plot
plt.figure(figsize=(7,5), dpi= 10)
fig, axes = joypy.joyplot(Total_MSW, column=['msw_per_day_per_person'], 
                          by="income_id", xlim='msw_per_day_per_person', figsize=(10,5), alpha=0.4)

# # Decoration
plt.title('Joy Plot of income in world', fontsize=12)
plt.xlabel('Waste produced per person per day (kgs)')
#plt.xlim(-2.0, 4.5)
plt.show()
```
