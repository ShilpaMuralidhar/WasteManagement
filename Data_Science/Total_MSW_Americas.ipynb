```python
%matplotlib notebook
import pandas as pd
pd.set_option('display.max_rows', None)
#from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
country_data = "/Users/shilpa/Desktop/Boringppl_project/WasteManagement/Data/country_level_data_0.csv"
country_data_df = pd.read_csv(country_data, encoding = 'utf-8')
country_data_df

Total_MSW = country_data_df[['country_name', 
                       'income_id', 
                       'gdp', 
                       'population_population_number_of_people', 
                       'total_msw_total_msw_generated_tons_year']]
Total_MSW.fillna(0).set_index('country_name')
```

```python
# rate of MSW generation per day = total_msw_total_msw_generated_tons_year / 365

#MSW['rate_per_day_kgs'] = MSW['total_msw_total_msw_generated_tons_year'] / 365
Total_MSW['msw_per_day'] =  Total_MSW['total_msw_total_msw_generated_tons_year'] / 365
Total_MSW['msw_per_day_per_person'] =  Total_MSW['total_msw_total_msw_generated_tons_year'] / Total_MSW["population_population_number_of_people"]
Total_MSW.head(1)

#Total_MSW['default_rank'] = Total_MSW['msw_per_day_per_person'].rank(ascending = True)
Total_MSW['default_rank'] = Total_MSW['msw_per_day_per_person']
Total_MSW.head(3)
```

```python
Northern_America = Total_MSW.loc[country_data_df['country_name'].
            isin(['Canada', 'Greenland', 'Mexico', 'United States'])].set_index('country_name')

Caribbean = Total_MSW.loc[country_data_df['country_name'] .\
            isin(['Anguilla', 'Antigua and Barbuda', 'Aruba', 'Bahamas', 'Barbados', 'Bermuda', 
                  'British Virgin Islands', 'Cayman Islands', 'Cuba', 'Curacao', 'Dominica', 'Dominican Republic',
                  'Grenada', 'Guadeloupe', 'Haiti', 'Jamaica', 'Martinique', 'Montserrat', 'Puerto Rico', 
                  'St. Kitts and Nevis', 'St. Lucia', 'St. Vincent and the Grenadines', 
                  'Trinidad and Tobago', 'Virgin Islands (U.S.)'])].set_index('country_name')


Central_America = Total_MSW.loc[country_data_df['country_name'] .\
             isin(['Belize', 'Costa Rica', 'El Salvador', 'Guetamala', 'Honduras', 'Nicaragua', 'Panama'])].set_index('country_name')


South_America = Total_MSW.loc[country_data_df['country_name'] .\
             isin(['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Columbia', 'Eucador', 'French Guiana(France)', 'Guyana',
                  'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuala'])].set_index('country_name')


Americas = [Northern_America, Caribbean, Central_America, South_America]
df_Americas = pd.concat(Americas)
df_Americas.drop('default_rank', axis=1)
```

```python
#x = df_Africa.loc[:, ['msw_per_day_per_person']]
#df_Africa['mpg_z'] = (x - x.mean())/x.std()
#df_Africa['colors'] = ['green' if x < 0.5 else 'red' for x in df_Africa['mpg_z']]
#df_Africa.sort_values('mpg_z', inplace=True)
#Total_MSW.reset_index(inplace=True)


x = df_Americas.loc[:, ['msw_per_day_per_person']]
df_Americas['colors'] = ['green' if x < 0.5 else 'red' for x in df_Americas['msw_per_day_per_person']]
df_Americas.sort_values('msw_per_day_per_person', inplace=True)
#Total_MSW.reset_index(inplace=True)


# Draw plot
plt.figure(figsize=(14,10), dpi= 80)
plt.hlines(y=df_Americas.index, xmin=0.01, xmax=df_Americas.msw_per_day_per_person, color=df_Americas.colors, alpha=0.4, linewidth=5)




#Decorations
plt.gca().set(ylabel='country_name', xlabel='msw_per_day_per_person')
plt.yticks(df_Americas.country_name, fontsize=12)
plt.title('Waste/person/day in America', fontdict={'size':10})
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
x = df_Americas.loc[:, ['msw_per_day_per_person']]
df_Americas['mpg_z'] = (x - x.mean())/x.std()
df_Americas['colors'] = ['green' if x < 0 else 'red' for x in df_Americas['mpg_z']]
df_Americas.sort_values('mpg_z', inplace=True)
#Total_MSW.reset_index(inplace=True)


#x = df_Africa.loc[:, ['msw_per_day_per_person']]
#df_Africa['colors'] = ['green' if x < 0.5 else 'red' for x in df_Africa['msw_per_day_per_person']]
#df_Africa.sort_values('msw_per_day_per_person', inplace=True)
#Total_MSW.reset_index(inplace=True)


# Draw plot
plt.figure(figsize=(14,10), dpi= 80)
plt.hlines(y=df_Americas.index, xmin=0, xmax=df_Americas.mpg_z, color=df_Americas.colors, alpha=0.4, linewidth=5)




#Decorations
plt.gca().set(ylabel='country_name', xlabel='mpg_z')
plt.yticks(df_Americas.country_name, fontsize=12)
plt.title('Waste/person/day in America', fontdict={'size':20})
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
x = df_Americas.loc[:, ['msw_per_day_per_person']]
df_Americas['mpg_z'] = (x - x.mean())/x.std()
df_Americas['colors'] = ['green' if x < 0 else 'red' for x in df_Americas['mpg_z']]
df_Americas.sort_values('mpg_z', inplace=True)
df_Americas.reset_index(inplace=True)

# Draw plot
plt.figure(figsize=(14,14), dpi= 80)
plt.hlines(y=df_Americas.country_name, xmin=0, xmax=df_Americas.mpg_z)
for x, y, tex in zip(df_Americas.mpg_z, df_Americas.index, df_Americas.mpg_z):
    t = plt.text(x, y, round(tex, 2), horizontalalignment='right' if x < 0 else 'left', 
                 verticalalignment='center', fontdict={'color':'green' if x < 0 else 'red', 'size':10})
    

# Decorations    
plt.yticks(df_Americas.country_name, fontsize=12)
plt.title('Population-Based Waste Generation Ranking in America', fontdict={'size':20})
plt.grid(linestyle='--', alpha=0.5)
plt.xlim(-2.0, 4.5)
plt.show()
```
