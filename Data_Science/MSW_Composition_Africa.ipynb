```python
%matplotlib notebook
import pandas as pd
#from IPython.display import display
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
country_data = "/Users/shilpa/Desktop/Boringppl_project/WasteManagement/Data/country_level_data_0.csv"
country_data_df = pd.read_csv(country_data, encoding = 'utf-8')
country_data_df.columns


```

```python
#Waste_Generated = pd.DataFrame(country_data_df, columns =[
    
waste = pd.DataFrame(country_data_df, columns=['country_name', 
                                'income_id',
                                'gdp', 
                                'population_population_number_of_people',
                                'composition_food_organic_waste_percent', 
                                'composition_glass_percent',
                                'composition_metal_percent', 
                                'composition_other_percent',
                                'composition_paper_cardboard_percent', 
                                'composition_plastic_percent',
                                'composition_rubber_leather_percent', 
                                'composition_wood_percent',
                                'composition_yard_garden_green_waste_percent',
                                'special_waste_agricultural_waste_tons_year',
                                'special_waste_construction_and_demolition_waste_tons_year',
                                'special_waste_e_waste_tons_year',
                                'special_waste_hazardous_waste_tons_year',
                                'special_waste_industrial_waste_tons_year',
                                'special_waste_medical_waste_tons_year',
                                'total_msw_total_msw_generated_tons_year'])

waste.head()
```

```python
All_waste = waste.rename(columns={'country_name': 'Country', 
                                'income_id': "Income",
                                'gdp': 'GDP', 
                                'population_population_number_of_people': 'Population',
                                'composition_food_organic_waste_percent': 'Organic', 
                                'composition_glass_percent': 'Glass',
                                'composition_metal_percent': 'Metal', 
                                'composition_other_percent': 'Other',
                                'composition_paper_cardboard_percent': 'Cardboard', 
                                'composition_plastic_percent': 'Plastic',
                                'composition_rubber_leather_percent': 'Rubber_Leather', 
                                'composition_wood_percent': 'Wood',
                                'composition_yard_garden_green_waste_percent': 'Yard_Garden',
                                'special_waste_agricultural_waste_tons_year': 'Agricultural',
                                'special_waste_construction_and_demolition_waste_tons_year': 'Building',
                                'special_waste_e_waste_tons_year': 'Electronic',
                                'special_waste_hazardous_waste_tons_year': 'Hazardous',
                                'special_waste_industrial_waste_tons_year': 'Industrial',
                                'special_waste_medical_waste_tons_year': 'Medical',
                                'total_msw_total_msw_generated_tons_year': 'Total_MSW'})

All_waste.fillna(0).head(3)
```

```python
General_waste = All_waste[['Country', 
                       'Income', 
                       'Organic', 
                        'Glass', 
                        'Metal', 
                        'Other', 
                        'Cardboard', 
                        'Plastic', 
                        'Rubber_Leather', 
                        'Wood', 
                        'Yard_Garden'
                        ]]
General_waste.fillna(0).set_index('Country').head(3)
```

```python
North_Africa = General_waste.loc[All_waste['Country'].isin(['Algeria', "Egypt, Arab Rep.", 
              "Libya", "Sudan", "Tunisia"])].set_index('Country')
North_Africa

West_Africa = General_waste.loc[All_waste['Country'] .\
            isin(['Benin', "Burkina Faso", "Cape Verde", "Côte d’Ivoire", "Gambia,The", "Ghana", 
                  "Guinea", "Guinea Bissau", "Liberia", "Mali", "Mauritania", "Niger", "Nigeria", 
                  "Senegal", "Sierra Leone","Togo"])].set_index('Country')

Central_Africa = General_waste.loc[All_waste['Country'] .\
            isin(['Angola', "Camaroon", "Central African Republic", "Chad", "Congo, Rep.", "Congo, Dem. Rep.", 
                  "Equatorial Guinea", "Gabon", "São Tomé and Príncipe"])].set_index('Country')

East_Africa = General_waste.loc[All_waste['Country'] .\
            isin(['Burundi', "Comaros", "Djibouti", "Eritrea", "Ethiopia", "Kenya", "Madagascar", "Malawi", 
                  "Mauritius", "Mozambique", "Rwanda", "Seychelles", "Somalia", "Somaliland", "South Sudan", 
                  "Tanzania", "Uganda", "Zambia", "Zimbabwe"])].set_index('Country')


Southern_Africa = General_waste.loc[All_waste['Country'] .\
            isin(['Botswana', "Lesotho", "Namibia", "South Africa", "Swaziland"])].set_index('Country')

Africa = [North_Africa, West_Africa, Central_Africa, East_Africa, Southern_Africa]
df_Africa = pd.concat(Africa)
df_Africa.fillna(0).head(3)
```

```python
# waste_generated_LIC = df_Africa.groupby('Income').get_group('UMC').fillna(0).drop_duplicates().reset_index('Country')
Africa_LIC_summary = df_Africa.sum(axis=0).fillna(0)
Africa_LIC_summary
```

```python
a =  pd.DataFrame(Africa_LIC_summary)
a.reset_index()
a.rename(columns={0:'Percent'}, inplace=True)
pd.options.display.float_format = '{:,.0f}'.format
a.index.names = ['Waste']
a.reset_index()
```

```python
waste_generated_LIC = df_Africa.groupby('Income').get_group('LIC').fillna(0).drop_duplicates().reset_index('Country')
Africa_LIC_summary = waste_generated_LIC.sum(axis=0).fillna(0)
Africa_LIC_summary





a =  pd.DataFrame(Africa_LIC_summary)
a.reset_index()
a.rename(columns={0:'Percent'}, inplace=True)
pd.options.display.float_format = '{:,.0f}'.format
a.index.names = ['Waste']
a.reset_index().drop([0, 1, 10])



```
